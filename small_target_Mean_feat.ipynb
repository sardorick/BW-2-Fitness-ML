```python
data_test = hdf.iloc[:25590, :]
data_train = hdf.iloc[25590: , :]
```

```python
X_train = data_train.iloc[:,:12]
y_train = data_train.iloc[:, 12:]

X_test = data_test.iloc[:, :12]
y_test = data_test.iloc[:, 12:]
```

```python
y_test
```

```python
tree_classifiers= {
  "Decision Tree": DecisionTreeClassifier(random_state=0),
  "Extra Trees":ExtraTreesClassifier(random_state=0),
  "Random Forest": RandomForestClassifier(random_state=0)
}
```

```python
results = pd.DataFrame({'Model': [], 'Accuracy': [], 'Bal Acc.': [], 'Time': []})

for model_name, model in tree_classifiers.items():
    
    start_time = time.time()
    model.fit(X_train, y_train)
    total_time = time.time() - start_time
        
    pred = model.predict(X_test)
    
    results = results.append({"Model":    model_name,
                              "Accuracy": accuracy_score(y_test, pred)*100,
                              "Bal Acc.": balanced_accuracy_score(y_test, pred)*100,
                              "Time":     total_time},
                              ignore_index=True)
                              
                              
results_ord = results.sort_values(by=['Accuracy'], ascending=False, ignore_index=True)
results_ord.index += 1 
results_ord.style.bar(subset=['Accuracy', 'Bal Acc.'], vmin=0, vmax=100, color='Blues')
```
