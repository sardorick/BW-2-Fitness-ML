```python
data_test = hdf.iloc[:25700, :]
data_train = hdf.iloc[25700: , :]
```

```python
X_train = data_train.iloc[:,:14]
y_train = data_train.iloc[:, 14:]

X_test = data_test.iloc[:, :14]
y_test = data_test.iloc[:, 14:]
```

```python
y_test.value_counts()
```

```python
tree_classifiers= {
  "GBM": GradientBoostingClassifier(),
  "Logistic regression": LogisticRegression(max_iter=1000),
  "Decision Tree": DecisionTreeClassifier(),
  "Extra Trees":ExtraTreesClassifier(),
  "Random Forest": RandomForestClassifier()
}
```

```python
results = pd.DataFrame({'Model': [], 'Accuracy': [], 'Bal Acc.': [], 'Time': []})

for model_name, model in tree_classifiers.items():
    
    start_time = time.time()
    model.fit(X_train, y_train)
    total_time = time.time() - start_time
        
    pred = model.predict(X_test)
    
    results = results.append({"Model":    model_name,
                              "Accuracy": accuracy_score(y_test, pred)*100,
                              "Bal Acc.": balanced_accuracy_score(y_test, pred)*100,
                              "Time":     total_time},
                              ignore_index=True)
                              
                              
results_ord = results.sort_values(by=['Accuracy'], ascending=False, ignore_index=True)
results_ord.index += 1 
results_ord.style.bar(subset=['Accuracy', 'Bal Acc.'], vmin=0, vmax=100, color='Blues')
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
hyper_params_gbm = {
    "n_estimators": [100, 200, 500],
    "max_depth": [3, 6, 9],
    "learning_rate": [0.001, 0.01, 0.1, 1]
}

hyper_params_rf = {
    "n_estimators": [100, 200, 500],
    "max_depth": [None, 3, 6, 9],
    "learning_rate": [0.001, 0.01, 0.1, 1],
    "min_samples_leaf": [0.1, 0.01, 1]
}

hyper_params_extra = {
    "n_estimators": [100, 200, 500],
    "max_depth": [3, 6, 9],
    "learning_rate": [0.001, 0.01, 0.1, 1],
    "gamma": [0.01, 0.1]
}

hyper_params_decision = {
    "n_estimators": [100, 200, 500],
    "learning_rate": [0.001, 0.01, 0.1, 1],
    "splitter": ["best", "random"],
    "min_samples_leaf": [1, 3, 5]
}
```

```python
GS = GridSearchCV(estimator=tree_classifiers.get('GBM'), param_grid=hyper_params_gbm, scoring='accuracy', cv=5)
GS.fit(X_train, y_train)
print(GS.best_params_, GS.best_score_)
```

```python
for model_name, model in tree_classifiers.items():
    if model_name == 'GBM':
        GS = GridSearchCV(estimator=tree_classifiers.get('GBM'), param_grid=hyper_params_gbm, scoring='accuracy', cv=5)
        GS.fit(X_train, y_train)
        print(GS.best_params_, GS.best_score_)
    elif model_name == 'Decision Tree':
        GS = GridSearchCV(estimator=tree_classifiers.get('Decision Tree'), param_grid=hyper_params_decision, scoring='accuracy', cv=5)
        GS.fit(X_train, y_train)
        print(GS.best_params_, GS.best_score_)
    elif model_name == 'Extra Trees':
        GS = GridSearchCV(estimator=tree_classifiers.get('Extra Trees'), param_grid=hyper_params_extra, scoring='accuracy', cv=5)
        GS.fit(X_train, y_train)
        print(GS.best_params_, GS.best_score_)
    elif model_name == 'Random Forest':
        GS = GridSearchCV(estimator=tree_classifiers.get('Random Forest'), param_grid=hyper_params_rf, scoring='accuracy', cv=5)
        GS.fit(X_train, y_train)
        print(GS.best_params_, GS.best_score_)
        
```
