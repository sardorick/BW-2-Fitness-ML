```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

This notebook performs EDA on the database. We would like to analyze the data we have.


```python
df =  pd.read_csv('data\dataset_halfSecondWindow.csv')
```

```python
df.info()
```

```python
df.describe()
```

```python
plt.figure(figsize=(25, 12))
df.isnull().sum().sort_values(ascending=True).head(30).plot(kind='barh')
plt.title('Parameters with highest number of missing values')
plt.savefig('null_params.png')
```

```python
df.target.value_counts()
```

```python
df.user.value_counts()
```

```python
df.head()
```

```python
plt.figure(figsize=(25, 12))
sb.countplot(df['target'])
plt.title('Balance of targets')
plt.savefig('target.png')
```

```python
plt.figure(figsize=(25, 12))
sb.countplot(df['user'])
plt.title('Users with the most data')
plt.savefig('users.png')
```

```python
df.shape
```

```python
df =  df.drop(columns=['sound#mean', 'sound#min', 'sound#max', 'speed#mean', 'speed#min', 'speed#max', 'speed#std'])
```

```python
df.columns
```

```python
#plotting nan values --> yellow is nan
plt.figure(figsize=(25, 12))
sb.heatmap(df.isnull(), yticklabels=False, cbar=False,cmap='viridis')
plt.title('Heatmap of NaN values')
plt.savefig(
    'nans.png'
)
```

```python
df.groupby('user')['target'].value_counts()
```

```python
plt.figure(figsize=(25, 12))
sb.countplot(x='user', hue='target',data=df.sort_values(by=['user']))
plt.legend(loc='upper right')
plt.title('Data from users per target')
plt.show()
plt.savefig('data_per_user.png')
```
