```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

This notebook performs EDA on the database. We would like to analyze the data we have.


```python
df =  pd.read_csv('five_sec')
```

```python
df.info()
```

```python
df.describe()
```

```python
plt.figure(figsize=(25, 12))
df.isnull().sum().sort_values(ascending=True).head(30).plot(kind='barh')
plt.title('Parameters with highest number of missing values')
plt.savefig('null_params.png')
```

```python
df.target.value_counts()
```

```python
df.user.value_counts()
```

```python
df.head()
```

```python
df_u1 = df[df['user'] == 'U1']
```

```python
plt.figure(figsize=(25, 12))
sb.countplot(df_u1['target'])
plt.title('Balance of targets of U1')
plt.savefig('U1 target.png')
```

```python
plt.figure(figsize=(25, 12))
sb.countplot(df['target'])
plt.title('Balance of targets')
plt.savefig('target.png')
```

```python
df = df[df.user != 'U1']
df = df.dropna(how='all')
df['user'].unique()
```

```python
plt.figure(figsize=(25, 12))
sb.countplot(df['target'])
plt.title('Balance of targets without U1')
plt.savefig('target_no_u1.png')
```

```python
plt.figure(figsize=(25, 12))
sb.countplot(df['user'].sort_values())
plt.title('Users with the most data')
plt.savefig('users.png')
```

```python
df.shape
```

```python
df =  df.drop(columns=['sound#mean', 'sound#min', 'sound#max', 'speed#mean', 'speed#min', 'speed#max', 'speed#std'])
```

```python
df.columns
```

```python
#plotting nan values --> yellow is nan
plt.figure(figsize=(25, 12))
sb.heatmap(df.isnull(), yticklabels=False, cbar=False,cmap='viridis')
plt.title('Heatmap of NaN values')
plt.savefig(
    'nans.png'
)
```

```python
df.groupby('user')['target'].value_counts()
```

```python
plt.figure(figsize=(25, 12))
sb.countplot(x='user', hue='target',data=df.sort_values(by=['user']))
plt.legend(loc='upper right')
plt.title('Data from users per target')
plt.show()
plt.savefig('data_per_user.png')
```

```python
plt.figure(figsize=(25, 12))
sb.scatterplot(x = df['android.sensor.step_counter#mean'], y=df['android.sensor.accelerometer#mean'], hue=df['target'])
plt.xlabel('Number of steps mean')
plt.ylabel('Accelerator speed mean')
plt.title('Number of steps taken by users relative to accelerometer speed')
plt.savefig('steps_accelerator.png')
```

```python
plt.figure(figsize=(25, 12))
sb.scatterplot(x = df['android.sensor.step_counter#mean'], y=df['android.sensor.gyroscope#mean'], hue=df['target'])
plt.xlabel('Number of steps mean')
plt.ylabel('Gyroscope mean')
plt.title('Number of steps taken by users relative to Gyroscope data')
plt.savefig('steps_gyroscope.png')
```

```python

plt.figure(figsize=(25, 12))
sb.scatterplot(x = df['android.sensor.accelerometer#mean'], y=df['android.sensor.gravity#mean'], hue=df['target'])
plt.xlabel('Accelerator')
plt.ylabel('Gravity')
plt.title('Accelerator sensor data relative to gravity sensor')
plt.savefig('accelerator_gravity.png')
```
